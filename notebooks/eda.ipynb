```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets
users = pd.read_csv("../data/raw/users.csv")
flights = pd.read_csv("../data/raw/flights.csv")
hotels = pd.read_csv("../data/raw/hotels.csv")

# Basic Info
print("Users Data")
print(users.info())
print(users.head())

print("\nFlights Data")
print(flights.info())
print(flights.head())

print("\nHotels Data")
print(hotels.info())
print(hotels.head())
```

```python
print(users.isnull().sum())
print(flights.isnull().sum())
print(hotels.isnull().sum())
```

```python
sns.histplot(flights['price'], kde=True)
plt.title("Flight Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(flights.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
flights['airline'].value_counts().plot(kind='bar')
plt.title("Top Airlines")
plt.show()
```

```python
sns.histplot(hotels['price'], kde=True)
plt.title("Hotel Price Distribution")
plt.show()
```
